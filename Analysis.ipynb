```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\ADMIN\Desktop\Data_analysis\Chess Analysis\Chess_Looker_Ready.csv")

bubble_df = df.groupby("Name", as_index=False).agg({
    "Age": "count",
    "ELO": "max",
    "Appearances": "sum"
})

bubble_df = bubble_df.sort_values(by="Age", ascending=False)

plt.figure(figsize=(10,6))

plt.scatter(
    bubble_df["Age"],
    bubble_df["ELO"],
    s=bubble_df["Appearances"]*35,
    alpha=0.6
)

plt.xlabel("Age Count")
plt.ylabel("Maximum ELO")
plt.title("Age Count vs Maximum ELO")

plt.grid(True)

plt.show()
```
